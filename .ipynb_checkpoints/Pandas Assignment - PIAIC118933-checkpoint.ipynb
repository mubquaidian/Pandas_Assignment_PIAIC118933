There is no output because all the data in dataset is not integer, 
So let us clean the data first then plot graphs

```python
us_census['Income'] = us_census['Income'].replace('[\$,]', '', regex=True)
us_census['Income'] = pd.to_numeric(us_census['Income'])
us_census
```

```python
split_genderpop = us_census.GenderPop.str.split("_")
us_census["Men"] = split_genderpop.str.get(0)
us_census["Women"] = split_genderpop.str.get(1)
us_census
```

```python
# now to remove M and F using reverse str to pick up last letters
us_census["Men"] = us_census.Men.str[:-1]
us_census["Women"] = us_census.Women.str[:-1]
us_census
```

```python
# after spliting GenderPop column into Men and Women, GenderPop column should be dropped
us_census = us_census.drop(['GenderPop'], axis =1)
us_census
```

```python
# let us remove % signs first a few columns 
us_census['Hispanic'] = us_census.Hispanic.str[:-1]
us_census['Hispanic'] = pd.to_numeric(us_census.Hispanic)
us_census['White'] = us_census.White.str[:-1]
us_census['White'] = pd.to_numeric(us_census.White)
us_census['Black'] = us_census.Black.str[:-1]
us_census['Black'] = pd.to_numeric(us_census.Black)
us_census['Native'] = us_census.Native.str[:-1]
us_census['Native'] = pd.to_numeric(us_census.Native)
us_census['Asian'] = us_census.Asian.str[:-1]
us_census['Asian'] = pd.to_numeric(us_census.Asian)
us_census['Pacific'] = us_census.Pacific.str[:-1]
us_census['Pacific'] = pd.to_numeric(us_census.Pacific)
us_census['Men'] = pd.to_numeric(us_census.Men)
us_census['Women'] = pd.to_numeric(us_census.Women)
```

```python
us_census
```

```python
# Now to fill empty spaces in data with .fillna() function
us_census = us_census.fillna(value={'Women': us_census.TotalPop - us_census.Men})
us_census
```

```python
# Now in Pacific there are still NaN which could be replace with mean values in Pacific
import numpy as np
x = us_census['Pacific'].mean()
us_census['Pacific'] = us_census['Pacific'].replace(np.NaN, value=x)
us_census
```

```python
# let find either there are duplicates in data
duplicates = us_census.duplicated(['State'])
print(duplicates.value_counts())
```

```python
# so there are 9 duplicates entries in data.
us_census = us_census.drop_duplicates()
us_census
```

```python
# now we have done all amost all the necessary manipulations in us_census data
# let us plot a scatter graph

plt.scatter(us_census['Women'], us_census['Income'])
plt.title('Scatter Graph between Women and Income')
plt.show()
```

# Part 2

Let us analyze Petal Power Inventory data

```python
import pandas as pd

# let us import Petal Power Inventory data

inventory = pd.read_csv('inventory.csv')
inventory
```

# Answer Customer Emails

```python
# let us Inspect the first 10 rows of inventory belong to staten_island

staten_island = inventory.head(10)
staten_island
```

```python
# A customer just emailed you asking what products are sold at your Staten Island location. 
# Select the column product_description from staten_island and save it to the variable product_request 

product_request = staten_island['product_description']
print(product_request)
```

```python
# Another customer emails to ask what types of seeds are sold at the Brooklyn location.
# Select all rows where location is equal to Brooklyn and product_type is equal to seeds 
# and save them to the variable seed_request

seed_request = inventory.product_description[(inventory.location == 'Brooklyn') & (inventory.product_type == 'seeds')]
seed_request
```

# Inventory

```python
# Add a column to inventory called in_stock 
# which is True if quantity is greater than 0 and False if quantity equals 0.

inventory['in_stock'] = inventory.quantity == 0
inventory
```

```python
# Petal Power wants to know how valuable their current inventory is.

# Create a column called total_value that is equal to price multiplied by quantity.

inventory['total_value'] = inventory['price'] * inventory['quantity']
inventory
```

```python
# The Marketing department wants a complete description of each product for their catalog.

# The following lambda function combines product_type and product_description into a single string:

combine_lambda = lambda row: \
    '{} - {}'.format(row.product_type, row.product_description)
```

```python
# Using combine_lambda, create a new column in inventory called full_description 
# that has the complete description of each product.

inventory['full_descrition'] = inventory.apply(combine_lambda, axis =1)
```
